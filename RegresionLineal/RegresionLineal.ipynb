```python
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import HistGradientBoostingRegressor

```

```python
df = sns.load_dataset(name='mpg')
```

```python
df.head()
```

```python
df.plot.scatter(x="horsepower",y="mpg")
```

```python
df["horsepower"].fillna(df["horsepower"].mean(), inplace=True)

# Entrenar el modelo
model = LinearRegression()
model.fit(df[["horsepower"]], df["mpg"])
```

```python
model.predict([[1],[10]])
```

```python
print(f"la linea que intersecta en :{model.intercept_} y tiene una pendiente de {model.coef_}")
```

```python
model.coef_ *df["horsepower"]+ model.intercept_
```

```python
pred=pd.Series(model.predict(df[["horsepower"]]))
```

```python
df["Prediccion"]=pred
```
